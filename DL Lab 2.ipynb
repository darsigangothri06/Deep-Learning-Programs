# 2) Implement Activation Functions in Neural Networks and analyse their usage.

```python
# 1)Identity Activation Function
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10,10) or np.linspace(-10,10)
y = x
plt.plot(x,y,c="green")
plt.title("Identity Activation Function")
plt.show()
```

```python
# 2)Binary Activation Function
import matplotlib.pyplot as plt
import numpy as np
y = []
x = np.arange(10,10) or np.linspace(-10,10)
for i in range(x.shape[0]):
  if(x[i]>0):
    y.append(1)
  else:
    y.append(0)
plt.plot(x,y,c="green")
plt.title("Binary Activation Function")
plt.show()
```

```python
# 3) Sigmoid Activation Function
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
# Implement Sigmoid Function
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
# 4) Tanh Activation Function
def tanh(x):
  return np.tanh(x)
```

```python
# Implement Tanh Function
x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :Tanh')
plt.show()
```

```python
# 5) Relu Activation Function
def RELU(x):
  x1=[]
  for i in x:
    if i<0:
      x1.append(0)
    else:
      x1.append(i)
    
  return x1
```

```python
# Implementing RELU Function
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :RELU')
plt.show()
```

```python
# 6) Softmax Activation Function
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x) , axis =0)
```

```python
# Implementing softmax() function
x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function :Softmax')
plt.show()
```
