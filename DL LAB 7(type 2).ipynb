#7) Train a simple Recurrent Neural Network using an Embedding layer and a Simple RNN layer for movie review classification problem.

```python
# Importing Packages
import tensorflow as tf
from tensorflow import keras
import numpy as np

(x_train, y_train), (x_test, y_test) = keras.datasets.imdb.load_data(num_words=10000)
```

```python
# Spliting Data into Train and Test
maxlen = 500
x_train = keras.preprocessing.sequence.pad_sequences(x_train, maxlen=maxlen)
x_test = keras.preprocessing.sequence.pad_sequences(x_test, maxlen=maxlen)

y_train = np.array(y_train, dtype=np.float32)
y_test = np.array(y_test, dtype=np.float32)
```

```python
# Building the Model
model = keras.models.Sequential([
    keras.layers.Embedding(10000, 128, input_length=maxlen),
    keras.layers.SimpleRNN(32),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
# COmpiling the Model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the Model and save in variable history
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

```python
# Printing Test Loss and Test Accuracy
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_acc)
```
